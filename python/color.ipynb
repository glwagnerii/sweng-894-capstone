## Colors for UI 

These colors are from the [Apple human interface guidelines](https://developer.apple.com/design/human-interface-guidelines/color)

```python
from coloraide import Color
import matplotlib.pyplot as plt
```

```python
def plot_named_colors(color_dict, facecolor):
    fig, ax = plt.subplots(figsize=(12, 3), facecolor=facecolor)
    ax.set_facecolor(facecolor)  # Set axes background too

    # Choose text color based on facecolor
    text_color = "white" if facecolor == "black" else "black"

    for idx, (name, rgba) in enumerate(color_dict.items()):
        rgba_values = rgba.replace('rgba(', '').replace(')', '').split(',')
        rgb = tuple(int(float(v)) for v in rgba_values[:3])
        alpha = float(rgba_values[3]) if len(rgba_values) == 4 else 1.0

        c = Color(f'rgb({rgb[0]} {rgb[1]} {rgb[2]})')
        oklch = c.convert('oklch').to_string()

        ax.add_patch(
            plt.Rectangle((idx, 0), 1, 1, color=(rgb[0]/255, rgb[1]/255, rgb[2]/255, alpha))
        )
        # Set text color
        ax.text(idx + 0.5, 1.05, f"{name}\n{rgba}", ha='center', va='bottom', fontsize=8, rotation=90, color=text_color)
        print(f"--cc_{name}: {oklch};")

    ax.set_xlim(0, len(color_dict))
    ax.set_ylim(0, 1.2)
    ax.axis('off')
    plt.tight_layout()
    plt.show()
```
